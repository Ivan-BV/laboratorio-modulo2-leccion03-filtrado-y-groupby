```python
# Carga el dataset de trabajo
import pandas as pd
import numpy as np

df = pd.read_csv("datos/netflix_merge.csv", index_col=0)

df.head()
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
condicion_fecha = df["release_year"] == 2021
condicion_genero = df["Genre"] == "Movie"
df_titulos = df[condicion_fecha][condicion_genero]
df_titulos.head()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
condicion_duracion = df["duration"] == "90 min"
df_duracion = df[condicion_duracion]
df_duracion[["title", "duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
condicion_pais = df["country"] == "India"
condicion_fecha = df["release_year"] > 2018
df_pais_fecha = df[condicion_pais][condicion_fecha]
df_pais_fecha["title"].reset_index()
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_clasificacion = df["rating"] == "TV-14"
condicion_clasificacion2 = df["rating"] == "R"

df[condicion_clasificacion | condicion_clasificacion2]["title"].reset_index()
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
condicion_directores = df["director"] == "Christopher Nolan"
condicion_directores2 = df["director"] == "Martin Scorsese"

df[condicion_directores | condicion_directores2][["title", "director"]].reset_index()

# df_prueba["director"].str.contains("^c", case=False, na=False)

#patron_regex = r"^christo"
#filtro = df_prueba["director"].str.contains(patron_regex, case=False, na=False)
#df_filtrado = df_prueba[filtro]
#df_filtrado.sort_values(by="director")
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
paises = ["United States", "United Kingdom", "Canada"]
condicion_lenguajes = df["Language"].value_counts()
df[df["country"].isin(paises)][["title", "country"]]
df["Language"].str.contains(r"\w*/\w*", case=False)
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_fecha = df["release_year"].between(2015, 2020, inclusive="both")
df[condicion_fecha][["title", "release_year"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_puntuacion = df["IMDB Score"].between(6.5, 8.0)
condicion_duracion = df["duration"].isnull()

df[condicion_puntuacion][condicion_duracion][["title", "duration"]].reset_index()
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
lista_duraciones = ["90 min", "150 min"]
condicion_duracion = df["duration"].isin(lista_duraciones)

df[condicion_duracion][["title", "duration"]]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
patron_regex = r"love"
condicion_titulo = df["title"].str.contains(patron_regex, case=False)
condicion_tipo = df["type"] == "TV Show"

df[condicion_titulo & condicion_tipo]["title"].reset_index()
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron_regex = r"johnson"
condicion_director = df["director"].str.contains(patron_regex, case=False, na=False)

df[condicion_director][["title", "director"]]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
condicion_tipo = df["listed_in"] == "Documentaries"
patron_regex = r"mike tyson"
condicion_cast = df["cast"].str.contains(patron_regex, case=False, na=False)

df[condicion_tipo & condicion_cast][["title", "listed_in", "cast"]]
```

#### Ejercicio 13: Selecciona solo las columnas 'title', 'country' y 'release_year' del dataset.

```python
df[["title", "country", "release_year"]]
```

#### Ejercicio 14: Filtra las columnas que contienen la palabra 'date'.

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
condicion_tipo = df["type"] == "Movie"
condicion_fecha = df["release_year"] < 2010

df[condicion_tipo & condicion_fecha][["title", "type", "release_year"]]

#df["release_year"].describe().reset_index()
```
